```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#plt.rcParams['figsize']=(20.0,10.0)

#Read the csv data
data = pd.read_csv('ex1data1.csv')
print(data.shape)
data.head()
```

```python
#collect x and y values
x= data['x_val'].values
y= data['y_val'].values
```

```python
print(x)
```

```python
print(y)
```

```python
cost_arr = []
iter_arr = []
def gradient_desc(x,y):
    m = b = 0
    iterations= 1000
    n = len(x)
    cost=0
    learning_rate = 0.008
    
    
    for i in range(iterations):
        predicted_y = m*x+b
        cost= (1/n)*sum([val**2 for val in (y-predicted_y)])
        dm= -(2/n)*sum(x*(y-predicted_y))
        db = -(2/n)*sum(y-predicted_y)
        m=m-learning_rate*dm
        b=b-learning_rate*db
        cost_arr.append(cost)
        iter_arr.append(i)
        print ("m {}, b {}, cost {} iteration {}".format(m,b,cost, i))
    return m, b
        
        
m,b = gradient_desc(x,y)
predicted_y = m*x + b
    
    
plt.scatter(x, y) 
plt.plot([min(x), max(x)], [min(predicted_y), max(predicted_y)], color='red')  # regression line
plt.show()




```

```python
plt.plot(iter_arr,cost_arr,color ='green')
```
